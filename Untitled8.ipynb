```python
import requests
from bs4 import BeautifulSoup as bs
base_url = 'https://is.signwiki.org/api.php'
```

```python
# data_url = 'https://is.signwiki.org/api.php?action=query&format=json&export=1&exportnowrap=1&generator=allpages&utf8=1&gapnamespace=0&gaplimit=20'
params = {
    'action': 'query',
    'format': 'json',
    'export': '1',
    'exportnowrap' : '1',
    'generator' : 'allpages',
    'apcontinue': 'Jóakim Aðalönd',
    'utf8' : '1',
    'gapnamespace' : '0',
    'gaplimit' : '20',
}

def get_data(page):
    title = page.find('title').text
    info = page.find('text').text.split('{{')[1].split('}}')[0].replace('\n','').split('|')
    info_type = info[0]
    info_dict = {k:v for k, _, v in [el.partition('=') for el in info[1:]]}
    
    output_dict = {
        'title':title,
        'info_type':info_type
    }
    output_dict.update(info_dict)
    return output_dict
```

```python
res = requests.get(base_url,params=params)
data = bs(res.text, "lxml")
pages = data.find_all('page')
```

```python
pages_data = [get_data(page) for page in pages]
' '.join([page['title'] for page in pages_data])
```

```python
# for result in query({'generator': 'allpages', 'prop': 'links'}):
#     # process result data
# ...
def query(request):
    request['action'] = 'query'
    request['format'] = 'json'
    lastContinue = {}
    while True:
        # Clone original request
        req = request.copy()
        # Modify it with the values returned in the 'continue' section of the last result.
        req.update(lastContinue)
        # Call API
        result = requests.get(base_url, params=req).json()

        if 'error' in result:
            raise Error(result['error'])
        if 'warnings' in result:
            print(result['warnings'])
        if 'query' in result:
            yield result['query']
        if 'continue' not in result:
            break
        lastContinue = result['continue']
```

```python
page_names = []
for result in query({
    'list': 'allpages',
    'aplimit': 'max'
    }):
    for page in result['allpages']:
        page_names.append(page['title'])

with open('page_names.txt','w') as f:
    f.write('\n'.join(page_names))
```
